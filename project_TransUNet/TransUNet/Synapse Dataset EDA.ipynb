```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import h5py
```

```python
base_path = '../data/Synapse/train_npz'
```

```python
npz_files_with_label = []
npz_files_without_label = []
max_label = -1
for f in os.listdir(base_path):
  x = np.load(f'{base_path}/{f}')
  img = x['image']
  label = x['label']
  if len(np.unique(label)) > 1:
    npz_files_with_label.append(f)
  else:
    npz_files_without_label.append(f)
  max_label = max(max(np.unique(label)), max_label)
#   if len(npz_files_with_label) > 5 and len(npz_files_without_label) > 5:
#     break
```
